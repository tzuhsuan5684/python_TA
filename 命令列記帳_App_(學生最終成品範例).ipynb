```
import csv
import os # å¼•å…¥ os æ¨¡çµ„ä¾†æª¢æŸ¥æª”æ¡ˆæ˜¯å¦å­˜åœ¨

# --- å…¨åŸŸè®Šæ•¸ ---
# å®šç¾©å„²å­˜è¨˜éŒ„çš„æª”æ¡ˆåç¨±
FILENAME = "records.csv"

# --- å‡½å¼å®šç¾© ---

def display_menu():
    """
    åŠŸèƒ½ï¼šå°å‡ºä¸€å€‹ç¾è§€ã€æ¸…æ™°çš„ä¸»é¸å–®ã€‚
    """
    print("\n" + "="*30)
    print("      ğŸ“ æ­¡è¿ä½¿ç”¨å‘½ä»¤åˆ—è¨˜å¸³æœ¬ ğŸ“")
    print("="*30)
    print("  1. æ–°å¢ä¸€ç­†æ¶ˆè²» (add)")
    print("  2. åˆ—å‡ºæ‰€æœ‰æ¶ˆè²» (list)")
    print("  3. è¨ˆç®—ç¸½æ”¯å‡º (total)")
    print("  4. é›¢é–‹ç¨‹å¼ (quit)")
    print("-"*30)

def load_records(filename):
    """
    åŠŸèƒ½ï¼šå¾æŒ‡å®šçš„ CSV æª”æ¡ˆè®€å–ç´€éŒ„ã€‚
    åƒæ•¸ï¼š
        filename (str): è¦è®€å–çš„æª”æ¡ˆåç¨±ã€‚
    å›å‚³ï¼š
        list: ä¸€å€‹åŒ…å«æ¶ˆè²»ç´€éŒ„å­—å…¸çš„æ¸…å–®ã€‚å¦‚æœæª”æ¡ˆä¸å­˜åœ¨ï¼Œå‰‡å›å‚³ç©ºæ¸…å–®ã€‚
    """
    expenses = []
    # æª¢æŸ¥æª”æ¡ˆæ˜¯å¦å­˜åœ¨
    if not os.path.exists(filename):
        return expenses # æª”æ¡ˆä¸å­˜åœ¨ï¼Œç›´æ¥å›å‚³ç©ºæ¸…å–®

    try:
        with open(filename, mode='r', encoding='utf-8', newline='') as file:
            reader = csv.DictReader(file)
            for row in reader:
                # å°‡é‡‘é¡å¾å­—ä¸²è½‰ç‚ºæµ®é»æ•¸
                row['amount'] = float(row['amount'])
                expenses.append(row)
        print(f"âœ… æˆåŠŸå¾ {filename} è¼‰å…¥ {len(expenses)} ç­†èˆŠç´€éŒ„ï¼")
    except Exception as e:
        print(f"âŒ è®€å–æª”æ¡ˆæ™‚ç™¼ç”ŸéŒ¯èª¤ï¼š{e}")
    return expenses

def save_records(filename, expenses):
    """
    åŠŸèƒ½ï¼šå°‡æ¶ˆè²»ç´€éŒ„æ¸…å–®å¯«å…¥æŒ‡å®šçš„ CSV æª”æ¡ˆã€‚
    åƒæ•¸ï¼š
        filename (str): è¦å¯«å…¥çš„æª”æ¡ˆåç¨±ã€‚
        expenses (list): åŒ…å«æ¶ˆè²»ç´€éŒ„å­—å…¸çš„æ¸…å–®ã€‚
    """
    try:
        with open(filename, mode='w', encoding='utf-8', newline='') as file:
            # å®šç¾© CSV æª”æ¡ˆçš„æ¬„ä½æ¨™é ­
            fieldnames = ['item', 'amount']
            writer = csv.DictWriter(file, fieldnames=fieldnames)

            writer.writeheader() # å¯«å…¥æ¨™é ­
            writer.writerows(expenses) # å¯«å…¥æ‰€æœ‰ç´€éŒ„
        print(f"ğŸ’¾ æ‰€æœ‰ç´€éŒ„å·²æˆåŠŸå„²å­˜è‡³ {filename}ï¼")
    except Exception as e:
        print(f"âŒ å„²å­˜æª”æ¡ˆæ™‚ç™¼ç”ŸéŒ¯èª¤ï¼š{e}")

def add_expense(expenses):
    """
    åŠŸèƒ½ï¼šå¼•å°ä½¿ç”¨è€…æ–°å¢ä¸€ç­†æ¶ˆè²»ç´€éŒ„ã€‚
    åƒæ•¸ï¼š
        expenses (list): è¦æ–°å¢ç´€éŒ„çš„æ¸…å–®ã€‚
    """
    print("\n--- ğŸ“ æ–°å¢æ¶ˆè²»ç´€éŒ„ ---")
    item = input("è«‹è¼¸å…¥æ¶ˆè²»å“é …ï¼š")

    # æŒçºŒè©¢å•ç›´åˆ°è¼¸å…¥æœ‰æ•ˆçš„é‡‘é¡
    while True:
        try:
            amount_str = input("è«‹è¼¸å…¥æ¶ˆè²»é‡‘é¡ï¼š")
            amount = float(amount_str)
            if amount > 0:
                break # é‡‘é¡æœ‰æ•ˆï¼Œè·³å‡ºè¿´åœˆ
            else:
                print("âŒ é‡‘é¡å¿…é ˆå¤§æ–¼ 0ï¼Œè«‹é‡æ–°è¼¸å…¥ã€‚")
        except ValueError:
            print("âŒ è¼¸å…¥ç„¡æ•ˆï¼Œè«‹è¼¸å…¥ä¸€å€‹æ•¸å­—ã€‚")

    # å»ºç«‹æ–°çš„æ¶ˆè²»å­—å…¸
    new_expense = {'item': item, 'amount': amount}
    expenses.append(new_expense)
    print(f"âœ… æˆåŠŸæ–°å¢ä¸€ç­†ç´€éŒ„ï¼š{item} - ${amount}")

def list_expenses(expenses):
    """
    åŠŸèƒ½ï¼šå°‡æ‰€æœ‰æ¶ˆè²»ç´€éŒ„ä»¥æ•´é½Šçš„æ ¼å¼å°å‡ºã€‚
    åƒæ•¸ï¼š
        expenses (list): åŒ…å«æ¶ˆè²»ç´€éŒ„çš„æ¸…å–®ã€‚
    """
    print("\n" + "-"*30)
    print("          ğŸ“œ æ¶ˆè²»åˆ—è¡¨ ğŸ“œ")
    print("-"*30)

    if not expenses:
        print("ç›®å‰æ²’æœ‰ä»»ä½•æ¶ˆè²»ç´€éŒ„...")
    else:
        # ä½¿ç”¨ f-string é€²è¡Œæ ¼å¼åŒ–å°é½Š
        # ^20 ä»£è¡¨ç½®ä¸­å°é½Šï¼Œå¯¬åº¦ç‚º 20
        # <10 ä»£è¡¨é å·¦å°é½Šï¼Œå¯¬åº¦ç‚º 10
        print(f"{'å“é …':<20} | {'é‡‘é¡':>10}")
        print("-"*32)
        for expense in expenses:
            print(f"{expense['item']:<20} | ${expense['amount']:>9.2f}")

    print("-"*30)

def calculate_total(expenses):
    """
    åŠŸèƒ½ï¼šè¨ˆç®—ä¸¦é¡¯ç¤ºç¸½æ”¯å‡ºã€‚
    åƒæ•¸ï¼š
        expenses (list): åŒ…å«æ¶ˆè²»ç´€éŒ„çš„æ¸…å–®ã€‚
    """
    total = 0
    for expense in expenses:
        total += expense['amount']

    print("\n--- ğŸ’° ç¸½æ”¯å‡ºçµ±è¨ˆ ---")
    print(f"æ‚¨ç›®å‰ç¸½å…±èŠ±è²»äº†ï¼š ${total:.2f}")
    print("--------------------")

def main():
    """
    åŠŸèƒ½ï¼šç¨‹å¼çš„ä¸»æ§åˆ¶ä¸­å¿ƒï¼ŒåŒ…å«ä¸»è¿´åœˆã€‚
    """
    # ç¨‹å¼å•Ÿå‹•æ™‚ï¼Œå…ˆè¼‰å…¥èˆŠç´€éŒ„
    all_expenses = load_records(FILENAME)

    while True:
        display_menu()
        command = input("è«‹è¼¸å…¥æŒ‡ä»¤ï¼š").lower() # å°‡è¼¸å…¥è½‰ç‚ºå°å¯«ï¼Œæ–¹ä¾¿æ¯”å°

        if command == 'quit' or command == '4':
            break
        elif command == 'add' or command == '1':
            add_expense(all_expenses)
        elif command == 'list' or command == '2':
            list_expenses(all_expenses)
        elif command == 'total' or command == '3':
            calculate_total(all_expenses)
        else:
            print("âŒ ç„¡æ³•è¾¨è­˜çš„æŒ‡ä»¤ï¼Œè«‹é‡æ–°è¼¸å…¥ã€‚")

    # ç¨‹å¼çµæŸå‰ï¼Œå„²å­˜æ‰€æœ‰ç´€éŒ„
    save_records(FILENAME, all_expenses)
    print("ğŸ‘‹ æ„Ÿè¬ä½¿ç”¨ï¼Œå†è¦‹ï¼")

# --- ç¨‹å¼é€²å…¥é» ---
# ç•¶é€™å€‹æª”æ¡ˆè¢«ç›´æ¥åŸ·è¡Œæ™‚ï¼Œæ‰åŸ·è¡Œ main() å‡½å¼
if __name__ == "__main__":
    main()
```
